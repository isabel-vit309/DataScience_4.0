#### Conceitos de Python
A atividade tem como objetivo que os alunos foquem na seleção e análise de um conjunto de dados educacionais disponibilizado pelo INEP (Instituto Nacional de Estudos e Pesquisas Educacionais Anísio Teixeira), através de seu portal de Indicadores Educacionais. O objetivo central é desenvolver competências em análise de dados com Python e visualização com Power BI, aplicando técnicas práticas aprendidas em aula.

Analisar os dados de distorção idade-série é importante porque mostra onde o fluxo escolar não está funcionando como deveria. Isso acontece quando o aluno está em uma série que não corresponde à sua idade, por exemplo, um adolescente ainda no ensino fundamental. Esse tipo de situação pode indicar que ele já repetiu de ano, entrou tarde na escola ou não teve o apoio necessário para acompanhar os estudos. Entender esses dados ajuda a pensar em soluções para evitar que os alunos fiquem para trás

A seguir vamos carregar o conjunto de dados (Inep - Taxa de Distorção idade-série) para responder a seguinte pergunta: A taxa de distorção idade-série é maior na área rural ou na área urbana?

OBS: A Taxa de Distorção idade-série traz as seguintes análises:

Taxa de Distorção de Idade-Série por Localização e Dependência Administrativa, 
Nos Níveis de Ensino Fundamental e Médio
-Brasil, Regiões Geográficas e Unidades da Federação - 2024

```python
import pandas as pd

tdi = pd.read_excel('datasets/TDI_BRASIL_REGIOES_UFS_2024.xlsx', skiprows=8)
tdi
```

```python
tdi.info()
```

```python
tdi.describe()
```

Todas as colunas presentes no arquivo estão sendo identificadas como Object, ou seja, estão sendo tratadas como String, por isso para seguir com a análise dos dados, iremos selecionar as colunas que seram necessárias para a analise desses dados 

As colunas selecionadas serão: Dependência Administrativa, Localização, Ensino Fundamental - Total e Ensino Médio - Total

```python
tdi = tdi.iloc[:, [1, 2, 3, 4, 16]]
tdi
```

A seguir, vamos renomear as colunas para facilitar a identificação e compreensão do que cada uma representa

UG = Unidade Geográfica
LOC = Localização
DA = Dependência Administrativa
TEF = Total Ensino Fundamental
TEM = Total Ensino Médio

```python
colunas = ["UG" , "LOC", "DA", "TEF", "TEM"]
tdi.columns = colunas
tdi
```

```python
tdi.info()
```

Mesmo após renomear as colunas, elas ainda aparecem como do tipo object, ou seja, estão sendo tratadas como strings. Para corrigir isso e permitir análises numéricas corretamente, vamos converter essas colunas para o tipo adequado utilizando o seguinte comando

```python
df = tdi.replace('--', 0)
```

```python
df.info()
```

Colocando a Média da Taxa de Distorção Idade-Série levando em consideração a Localização (Rural ou Urbana) 
junto com o Total do Ensino Fundamental e Total do Ensino Médio

```python
df.groupby("LOC")[["TEF", "TEM"]].median()
```

Colocando a Média da Taxa de Distorção Idade-Série levando em consideração a Unidade Geográfica (Estado) e a Localização (Rural ou Urbana)
junto com o Total do Ensino Fundamental e Total do Ensino Médio

```python
df.groupby(["UG", "LOC"])[["TEF", "TEM"]].median()
```

Para realizar uma análise mais detalhada, vamos calcular a média da Taxa de Distorção Idade-Série em cada estado e em cada tipo de rede de ensino (municipal, estadual, federal, privada, entre outras). Essa abordagem permite identificar onde esse problema é mais recorrente e em quais redes ele é mais grave, auxiliando no direcionamento de políticas educacionais mais eficientes

```python
pd.pivot_table(df, values = "TEF",  index = "UG", columns= "DA", aggfunc="median")
```

# Visualizando o Gráfico de Dados 

Nesse gráfico iremos conseguir a resposta para a pergunta feita no início da análise de dados (A Taxa de Distorção Idade-Série é maior na área rural ou área urbana?)

Gráfico em Barras

```python
import matplotlib.pyplot as plt 

df.groupby("LOC")[["TEF", "TEM"]].median().plot(kind="barh", figsize= (10, 5))
plt.title("Média de Taxa de Distoção de Idade-Série por Localização")
plt.xlabel("Quantidade média por Localização")
plt.tight_layout()
plt.show(   )
```

Gráfico em Colunas

```python
df.groupby("LOC")[["TEF", "TEM"]].median().plot(kind="bar", figsize=(8,5))
plt.title("Média de Taxa de Distorção Idade-Série por Localização")
plt.ylabel("Média de Taxa de Distorção Idade-Série")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
